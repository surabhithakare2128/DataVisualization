```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel = pd.read_csv('fuel-econ.csv')
print(fuel.shape)
fuel.head(2)
```

```python
#creating subset

np.random.seed(2018)
sample = np.random.choice(fuel.shape[0], 200, replace = False)
fuel_subset = fuel.loc[sample] #.loc tells that the index is as per the location
```

```python
#taking only transmission name from trans column
fuel_subset['trans_type'] = fuel['trans'].apply(lambda x: x.split()[0])

g = sb.FacetGrid(data = fuel_subset, hue= 'trans_type', 
                 hue_order=['Automatic', 'Manual'], size = 4, aspect = 1.5) 
g.map(sb.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False)
g.add_legend()
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');

#hue_order: defines the order of point to be plotted
#size: changes the size of visualization to 4
#aspect: makes the lenght 1.5 times bigger than height

#default palette works well because the datatype of hue is nominal
#following palette can be used when the datatype is ordinal eg "VClass"
```

```python
g = sb.FacetGrid(data = fuel_subset, hue= 'VClass', 
                 size = 4, aspect = 1.5) 
g.map(sb.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False)
g.add_legend()
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');

# here as the data is ordinal the color palette doesn't show increasing order
```

```python
g = sb.FacetGrid(data = fuel_subset, hue= 'VClass', 
                 size = 4, aspect = 1.5, palette='viridis_r') 
g.map(sb.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False)
g.add_legend()
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');

#but this can't be used for columns like Co2 becuase python will alot separate color for every value as shown below
```

```python
g = sb.FacetGrid(data = fuel_subset, hue= 'co2', 
                 size = 4, aspect = 1.5, palette='viridis_r') 
g.map(sb.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False)
g.add_legend()
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

```python
#to overcome above case, use scatter plot 

plt.scatter(data=fuel_subset, x = 'displ', y = 'comb', c = 'co2',
           cmap = 'viridis_r')
#c: is hue
#cmap: is used to set color palette
plt.colorbar(label='CO2 (g/mi)')
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');

#colorbar show colorbar at the side
```

```python
sb.pointplot(data = fuel, x ='VClass', y = 'comb', ci = 'sd', 
             hue='trans_type', linestyles="")  
#linestyles= ""(empty string) will remove lines
#errwidth = 0 will remove the error whiskers
plt.xticks(rotation = 20)
plt.ylabel('AVG Combined fuel effeciency ')

#since the levels are overlapping you can use dodge=true to avoid overlapping
```

```python
sb.pointplot(data = fuel, x ='VClass', y = 'comb', ci = 'sd', 
             hue='trans_type', linestyles="", dodge=True) 
plt.xticks(rotation = 20)
plt.ylabel('AVG Combined fuel effeciency ')

#you can change the pointplot to barchart to get an adapted clustered barchart
```

```python
sb.barplot(data = fuel, x ='VClass', y = 'comb', ci = 'sd', 
             hue='trans_type') 
plt.xticks(rotation = 20)
plt.ylabel('AVG Combined fuel effeciency ')
```

```python
#you can take one step further by using a boxplot for summarization

sb.boxplot(data = fuel, x ='VClass', y = 'comb', hue='trans_type') 
plt.xticks(rotation = 20)
plt.ylabel('AVG Combined fuel effeciency ')
```

```python
#you can also use heatmap to see color based on mean of third variable instead of color based on count

bins_x = np.arange(0.6,7+0.3, 0.3)
bins_y = np.arange(12.8,58+3, 3)

#generating plot using new bins
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin= 0.5, 
           cmap= 'viridis_r', bins = [bins_x,bins_y])
plt.colorbar() 
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
#cut function is used to find in which bin each point falls

# count number of points in each bin
displ_bins = pd.cut(fuel['displ'], bins_x, right = False,
                    include_lowest = True, labels = False).astype(int)
#astype(int) is added later to convert the values to int
comb_bins = pd.cut(fuel['comb'], bins_y, right = False,
                    include_lowest = True, labels = False).astype(int)
#print(disp_bins, comb_bins)

#counting numbers that fall in each bin
n_points = fuel.groupby([disp_bins, comb_bins]).size()
#print(n_points)

#but these multi-index series is kind of hard to wirk with
#so well will use reset_index and pivot to put series in array form

n_points = n_points.reset_index().pivot(index='displ', columns='comb').values
#print(n_points)

#now we can get all out point weights using these values
#we take co2 series and divide each element by the number points in each bin using cut bin vectors for indexing
#into the right place

co2_weights = fuel['co2'] / n_points[displ_bins,comb_bins]
#print(co2_weights)

#now we add co2_weights to weight parameter and plot the graph
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin= 0.5, 
           cmap= 'viridis_r', bins = [bins_x,bins_y], weights=co2_weights)
plt.colorbar() 
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```
