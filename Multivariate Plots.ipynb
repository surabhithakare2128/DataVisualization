```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel = pd.read_csv('fuel-econ.csv')
print(fuel.shape)
fuel.head(2)
```

```python
sb.regplot(data = fuel, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False)
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

```python
#but to see a less clumpsy plot let's take random 200 subset

np.random.seed(2018)
sample = np.random.choice(fuel.shape[0], 200, replace = False)
fuel_subset = fuel.loc[sample] #.loc tells that the index is as per the location
 
sb.regplot(data = fuel_subset, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False)
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

```python
#both regplot and scatterplot takes only one argument to in marker
#so we need to create create list that will pass point of observation to marker

#creating markers
ttype_markers = [['Automatic', 'o'],
                ['Manual', '^']]

#taking only transmission name from trans column
fuel['trans_type'] = fuel['trans'].apply(lambda x: x.split()[0])

#creating markers using trans_type
for ttype, marker in ttype_markers:
    plot_data = fuel_subset.loc[fuel_subset['trans_type'] == ttype]
    sb.regplot(data = plot_data, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False, marker = marker)
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
plt.legend(['Automatic', 'Manual'])
#adds legend to the visualization
```

```python
sb.regplot(data = fuel_subset, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False, 
          scatter_kws= {'s': fuel_subset['co2']})
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');

#this plot is god but the size of bubbles is to big.
#you canreduce it using following code
```

```python
sb.regplot(data = fuel_subset, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False, 
          scatter_kws= {'s': fuel_subset['co2']/4}) #reduces bubble size
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

```python
sb.regplot(data = fuel_subset, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False, 
          scatter_kws= {'s': fuel_subset['co2']/2}) #reduces bubble size
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

```python
#but the sizes don't tell you the quantity 
#You can do this in following way

#selecting size for legend 
sizes = [125,250,500]

#setting base color so whole plot will have same color
base_color = sb.color_palette()[0]

#creating empty legend object that will store all legends
legend_obj = [] 

#adding legends as per size to legend object
for s in sizes:
    legend_obj.append(plt.scatter([],[], s=s/2, color = base_color)) 
    #as x and y are empty lists so no points are plotted
    #and s= s/2 because that is the measure we have used in our program for plotting visual
    
#plotting legend
plt.legend([sizes]) #when there is only one parameter it's the label for plot objects

plt.legend(legend_obj, sizes, title = 'CO2 (g/mi)') 
#when there are two parameters, first is the list to objects to be plotted and second is the label
#title can be anything that describes the legend. Here it's the measurement of co2

#plotting graph
sb.regplot(data = fuel_subset, x = 'displ', y = 'comb', x_jitter = 0.04, fit_reg = False, 
          scatter_kws= {'s': fuel_subset['co2']/2}) #reduces bubble size
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```
