```python
datas = pd.read_csv("pokemon.csv")
```

```python
datas.head(10)
```

```python
sns.set()
sns.scatterplot(x="generation_id", y="height", data=datas);
```

```python
sns.set()
sns.relplot(x="weight", y="height", col="generation_id",hue="type_1", 
            style="generation_id", size="speed", data=datas); 
#default is scatter
#the following graph shows scatter plot for each generation id as per the height and weight
#hue = type_1 i.e clor as per type_1
#style = generation_id i.e. shape chnages as per id
```

```python
sns.set()
sns.relplot(x="special-attack", y="special-defense", col="generation_id", size="generation_id",
            hue = 'generation_id', kind = 'line', legend='full', data=datas); 
#legend = 'full' shows full legend
```

```python
sns.set()
sns.lmplot(x="special-attack", y="special-defense", col="generation_id",
            hue = 'generation_id', data=datas); 
```

```python
sns.catplot(x="generation_id", y="weight", hue="type_1",
            kind="swarm", data=datas);
#if there would have been gender then setting hue= 'gender' would have showed males and females in each category
```

```python
sns.catplot(x="generation_id", y="weight", hue="type_1",
            kind="violin", data=datas);
```

```python
sns.catplot(x="generation_id", y="weight", hue="generation_id",
            kind="box", data=datas);
```

```python
f, axes = plt.subplots(1, 3, sharey=True, figsize=(10, 3)) #1 is number 0f rows and 3 is number of columns
#sharey allows to share properties among different plots
sns.scatterplot(x="special-attack", y="special-defense", hue="generation_id", 
                legend = 'full', data=datas,  ax=axes[0])
sns.boxplot(x="generation_id", y="weight", hue="generation_id", data=datas,  ax=axes[1])
sns.lineplot(x="height", y="weight", hue="generation_id", data=datas,  ax=axes[2])
```

```python
sns.pairplot(data=datas, hue = 'generation_id');
```

```python
sns.scatterplot(x="generation_id", y="height", data=datas);
```

```python
#modifying above plot

sns.scatterplot(x="generation_id", y="height",
            hue="generation_id",
            palette="YlGnBu", markers=["D", "o"], sizes=(100, 500),
            edgecolor=".2", linewidth=.5, alpha=.75,
            data=datas);
```

```python
f, axes = plt.subplots(2, 2, sharey=True, figsize=(10,3)) #1 is number 0f rows and 3 is number of columns
#sharey allows to share properties among different plots
sns.scatterplot(x="special-attack", y="special-defense", hue="generation_id", 
                legend = 'full', data=datas,  ax=axes[0,0])
sns.boxplot(x="generation_id", y="weight", hue="generation_id", data=datas,  ax=axes[0,1])
sns.lineplot(x="height", y="weight", hue="generation_id", data=datas,  ax=axes[1,0])
```
