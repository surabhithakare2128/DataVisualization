```python
def log_trans(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)
```

```python
#so the appropriate bin size for displ is 0.3 and for comb is 3

#setting bin size for x
bins_x = np.arange(0,7+0.3, 0.3)

#setting bin size for y
bins_y = np.arange(0,58+3, 3)

#generating plot using new bins
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin= 0.5, 
           cmap= 'viridis_r', bins = [bins_x,bins_y])
plt.colorbar() 
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
#Creating bins from starting point of data i.e minimum

#setting bin size for x
bins_x = np.arange(0.6,7+0.3, 0.3)

#setting bin size for y
bins_y = np.arange(12.8,58+3, 3)

#generating plot using new bins
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin= 0.5, 
           cmap= 'viridis_r', bins = [bins_x,bins_y])
plt.colorbar() 
plt.xlabel('Displacement 1')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
#adding numbers on each box

#hist2d returns a number of different variables, including an array of counts
bins_x = np.arange(0.6,7+0.3, 0.3)
bins_y = np.arange(12.8,58+3, 3)

h2d = plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin= 0.5, 
                 cmap= 'viridis_r', bins = [bins_x,bins_y])
counts = h2d[0]

# loop through the cell counts and add text annotations for each
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 250: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'black')
```

```python
fuel['trans_type'] = fuel['trans'].apply(lambda x: x.split()[0])
fuel['trans_type']
```

```python
ct_counts = fuel.groupby(['VClass', 'trans_type']).size()#this gives combination of car class with their type 
#as pandas series 
print(ct_counts)

ct_counts = ct_counts.reset_index(name = 'count') #reset_index helps to convert series to dataframe

ct_counts = ct_counts.pivot(index='VClass', columns = 'trans_type', values = 'count') 
type(ct_counts)
#pivot is used to rearrange the data

sb.heatmap(ct_counts) #plotting heatmap
```

```python
#above heatmap is difficult to understand so we are adding annot that will display numbers
sb.heatmap(ct_counts, annot=True)
```

```python
#pointplot is used for qualitative vs quantitative, errorbar is used for quantitative vs quantitative
#errorba can be used if our x-variable is sorted and have only one y value for each x value
plt.errorbar(data = fuel, x ='displ', y = 'comb') 
plt.xlabel('Displacement')
plt.ylabel('AVG Combined fuel effeciency ')
```

```python
#setting data to make errorbar better

#setting bin values for displacement
bins_e = np.arange(0.6, 7+0.2, 0.2)
bins_c = bins_e[:-1] + 0.1 #this gives center values so centers are plotted in their accurate positions

#cut function is used to figure out which bin each data point should be used in
displ_binned = pd.cut(fuel['displ'], bins_e, include_lowest=True)
comb_mean = fuel['comb'].groupby(displ_binned).mean() 
#group by will group by displacement bins
#and take mean of each point that fall in that bin

#plotting errorbar
plt.errorbar(x = bins_c, y = comb_mean) 
plt.xlabel('Displacement')
plt.ylabel('AVG Combined fuel effeciency ')
#the plot shows general decrease in fuel efficiency by engine size

#there is a gap in line plot where there is no data
```

```python
#instead of taking mean we can also take standard deviation of whole dataset

bins_e = np.arange(0.6, 7+0.2, 0.2)
bins_c = bins_e[:-1] + 0.1 #this gives center values so centers are plotted in their accurate positions

#cut function is used to figure out which bin each data point should be used in
displ_binned = pd.cut(fuel['displ'], bins_e, include_lowest=True)
comb_std = fuel['comb'].groupby(displ_binned).std() 


#plotting errorbar
plt.errorbar(x = bins_c, y = comb_std) 
plt.xlabel('Displacement')
plt.ylabel('AVG Combined fuel effeciency ')
```

```python

bins_e = np.arange(0.6, 7+0.2, 0.2)
bins_c = bins_e[:-1] + 0.1 #this gives center values so centers are plotted in their accurate positions

#cut function is used to figure out which bin each data point should be used in
displ_binned = pd.cut(fuel['displ'], bins_e, include_lowest=True)
comb_std = fuel['comb'].groupby(displ_binned).std() 


#plotting errorbar
plt.errorbar(x = bins_c, y = comb_mean, yerr=comb_std) 
plt.xlabel('Displacement')
#this is rugplotplt.ylabel('AVG Combined fuel effeciency ')
```
